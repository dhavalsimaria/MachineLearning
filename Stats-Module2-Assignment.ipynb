```python
#Answer 1:
import numpy as np
z = np.random.random((10,2))
x,y = z[:,0], z[:,1]
r = np.sqrt(x**2 + y**2)
t = np.arctan2(y,x)
print(r)
print(t)
```

```python
#Answer 2:
import numpy as np
x = np.random.random(50)
print("Original Vector: ")
print(x)
x[x.argmax()] = 0
print("Maximum value replaced by 0: ")
print(x)
x[x.argmin()] = 100
print("Minimum value replaced by 100: ")
print(x)
```
