```python
import pandas as pd
import numpy as np
import csv
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Load train data
df_train = pd.read_csv("D:\\APP\\NITW\\Graphical Models\\Assignment-9\\Problem-1\\train.csv",sep=",")
```

```python
#Load test data
df_test = pd.read_csv("D:\\APP\\NITW\\Graphical Models\\Assignment-9\\Problem-1\\test.csv",sep=",")
```

```python
df_train = df_train.drop(columns=['Unnamed: 0', 'id', 'Unnamed: 32'], axis=1)
```

```python
df_test = df_test.drop(columns=['Unnamed: 0', 'id', 'Unnamed: 32'], axis=1)
```

```python
df_train_encoded = df_train.copy()
df_test_encoded = df_test.copy()
```

```python
le = preprocessing.LabelEncoder()
```

```python
df_train_encoded['diagnosis'] = le.fit_transform(df_train_encoded['diagnosis'])
df_test_encoded['diagnosis'] = le.fit_transform(df_test_encoded['diagnosis'])
```

```python
X_train = df_train_encoded.drop(columns=['diagnosis'])
y_train = df_train_encoded['diagnosis']
```

```python
X_test = df_test_encoded.drop(columns=['diagnosis'])
y_test = df_test_encoded['diagnosis']
```

```python
mmScaler = MinMaxScaler()
mmScaler.fit(X_train)
X_train[X_train.columns] = mmScaler.transform(X_train[X_train.columns])
X_test[X_test.columns] = mmScaler.transform(X_test[X_test.columns])
```

**Naive Bayes Classifier**

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
nb_accuracy = np.round(metrics.accuracy_score(y_test, y_pred), 4)
print("Gaussian Naive Bayes model accuracy:", nb_accuracy)
```

**Logistic Regression Model**

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
#lr.score(X_test, y_test)
lr_accuracy = np.round(metrics.accuracy_score(y_test, y_pred), 4)
print("Logistic Regression model accuracy:", lr_accuracy)
```

```python
#write your output in csv in this form(exactly to three decimals points)
data=[nb_accuracy, lr_accuracy]
result=pd.DataFrame(data)
result.to_csv('D:\\APP\\NITW\\Graphical Models\\Assignment-9\\Problem-1\\output.csv', header=False, index=False)
```
