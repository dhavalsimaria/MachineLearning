```python
#GROUPBY

df1 = pd.DataFrame({'SalesRep':['A', 'B', 'C', 'A', 'B', 'C'],
                    'Region':['North', 'South', 'North', 'South', 'East', 'South'],
                    'Qty':[10, 5, 8, 20, 23, 30],
                    'UnitPrice':[5, 10, 20, 6, 7, 8]})
```

```python
df1
```

```python
#Groupby - aggregating the data at some level [Categorical]
# YOU GROUPBY A CATEGORICAL VARIABLE -- U WANT TO LOOK AT AGGREGATE FOR A NUMERICAL VARIABLE

# Q1 - What is the avg quantity sold per region?
grouped_by_Result = df1.groupby(['Region']).mean()
```

```python
grouped_by_Result.columns
```

```python
# Indexes are row label - e.g.: East, North, South
#Q2 - Avg quantity of sold per salesrep per region
grouped_by_Result_salesrep_region = df1.groupby(['SalesRep','Region']).mean()
```

```python
grouped_by_Result_salesrep_region
```

```python
#Q3 - Sum of all quantities and avg unit price per salesrep
df1.groupby(['SalesRep']).agg({'Qty':'sum', 'UnitPrice':'mean'})
```

```python
#Q4 - For all regions which salesrep sold how many quantities
df1.groupby(['Region', 'SalesRep']).sum()
```

```python
# JOIN
df2 = pd.DataFrame({'SalesRep':['A', 'B', 'C', 'D'],
                    'City':['Mumbai', 'Bangalore', 'Delhi', 'Kolkata']})
df3 = pd.DataFrame({'SalesRep':['A', 'B', 'C', 'E'],
                    'QtySold':[10, 20, 30, 40]})
```

```python
df2
```

```python
df3
```

```python
# INNER JOIN
pd.merge(df2, df3, how = 'inner', on = 'SalesRep')
```

```python
# LEFT JOIN
pd.merge(df2, df3, how = 'left', on = 'SalesRep')
```

```python
# RIGHT JOIN
pd.merge(df2, df3, how = 'right', on = 'SalesRep')
```

```python
# OUTER JOIN
pd.merge(df2, df3, how = 'outer', on = 'SalesRep')
```

```python
salaries_data = pd.read_csv("D:\\APP\\NITW\\Applications using Python\\Module-2\\Demo2\\Salaries.csv")
```

```python
salaries_data.head()
```

```python
#Q-1: How did the cost for TotalPayBenefits change year-on-year?
salaries_data[['TotalPayBenefits', 'Year']].groupby('Year').sum()
```

```python
#Q-2: What are the 5 most frequent job titles in 2013 and how much they cost to the company?
salaries_data_2013 = salaries_data[salaries_data.Year == 2013]
```

```python
top_5_jobtitles = salaries_data_2013['JobTitle'].value_counts().head(5)
```

```python
top_5_jobtitles.index
```

```python
salaries_data_2013[salaries_data_2013['JobTitle'].isin(list(top_5_jobtitles.index))]
```

```python
salaries_data_2013[salaries_data_2013['JobTitle'].isin(list(top_5_jobtitles.index))]['TotalPayBenefits'].sum()
```

```python
# Who is the highest paid employee over all the year?
salaries_data.groupby(['EmployeeName']).agg({'TotalPayBenefits':'sum'}).sort_values(by = 'TotalPayBenefits', ascending = False).head(1)
```
