```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
#Generate numbers from 1 to 10 spaced by 0.3
x = np.arange(1, 10, 0.3)
y = [x1**2 for x1 in x]
y1 = [x1+10 for x1 in x]
y2 = [x1**3 for x1 in x]
```

```python
plt.scatter(x, y, color="y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Relationship b/w X and Y")
```

```python
plt.plot(x,y,color="g")
```

```python
# multiline plot
# When you are comparing the trend of 3 different product lines
plt.figure(figsize=(10, 10))
plt.plot(x, y, label="Square of X")
plt.plot(x, y1, label="X + 10")
plt.plot(x, y2, label="Cube of X")
plt.legend()
plt.grid(True)
plt.savefig("ABC")
```
